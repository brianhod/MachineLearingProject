# Encoder

## Date/Lag/Avr-Encoder

**Problemstellung:**
Zeitvariablen wie Monate oder Tage der Woche haben eine zyklische Natur (z.B. Januar folgt auf Dezember). Klassische maschinelle Lernmodelle wie Support Vector Regression (SVR) erkennen diese zyklischen Muster jedoch nicht automatisch. Zudem reichen einfache Zeitreihen nicht aus, um saisonale Schwankungen oder zeitliche Abhängigkeiten korrekt abzubilden.

**Lösung:**
Um die zyklische Struktur der Zeit und historische Preisinformationen korrekt zu berücksichtigen, wurden folgende Spalten erstellt:

*Zyklische Kodierung (Monat)*: Durch Umwandlung der Monate in Sinus- und Cosinus-Werte wird die zyklische Natur korrekt erfasst. Dies sorgt dafür, dass Januar und Dezember als benachbart angesehen werden.

*Lag-Features (Verzögerungsmerkmale)*: Verzögerungsmerkmale (z.B. Preise vor einer Woche oder einem Monat) helfen dem Modell, saisonale oder zeitliche Abhängigkeiten besser zu berücksichtigen.

*Gleitende Durchschnitte*: Kurzfristige Trends und saisonale Muster werden durch gleitende Durchschnitte geglättet, um das Modell robuster gegen kurzfristige Schwankungen zu machen.

Diese Features verbessern die Fähigkeit des Modells, zeitliche Abhängigkeiten und saisonale Muster korrekt zu erfassen.

```python
# Patrick DateEncoder für Monate
import numpy as np

# -------------------------------
# Extrahiere Zeitkomponenten
# -------------------------------
# Monat aus der "date"-Spalte extrahieren
df['month'] = df['date'].dt.month   # Monat extrahieren

# -------------------------------
# Zyklische Kodierung
# -------------------------------
# Monat zyklisch mit sin und cos kodieren, um die zyklische Natur (Januar -> Dezember) zu berücksichtigen
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

# -------------------------------
# Lag-Features (Verzögerungsmerkmale)
# -------------------------------
# Erstelle Verzögerungsmerkmale für den durchschnittlichen Preis
# Preis vor einer Woche und vor einem Monat
df['lag_7'] = df['average_price'].shift(7)   # Preis vor einer Woche
df['lag_30'] = df['average_price'].shift(30) # Preis vor einem Monat

# -------------------------------
# Gleitende Durchschnitte
# -------------------------------
# Gleitender Durchschnitt des Preises der letzten 7 Tage (Fenstergröße 14 für Glättung)
#df['moving_avg_7'] = df['average_price'].rolling(window=14).mean()

# Gleitender Durchschnitt des Preises der letzten 30 Tage
df['moving_avg_30'] = df['average_price'].rolling(window=30).mean()

# Auffüllen der NaN-Werte in den Lag- und Moving-Average-Spalten mit dem Wert der 'average_price' in der gleichen Zeile

# Fülle Lag-Features mit dem Wert der 'average_price' in der gleichen Zeile
df['lag_7'].fillna(df['average_price'], inplace=True)
df['lag_30'].fillna(df['average_price'], inplace=True)

# Fülle gleitende Durchschnitte mit dem Wert der 'average_price' in der gleichen Zeile
#df['moving_avg_7'].fillna(df['average_price'], inplace=True)
df['moving_avg_30'].fillna(df['average_price'], inplace=True)

# Entfernen der angegebenen Spalten
df = df.drop(columns=['month'])
```

```python
# Patrick GeoDistanceEncoder für geography

# Importieren der notwendigen Bibliotheken
from geopy.distance import geodesic

# Definieren der Städte mit ihren Koordinaten
static_coordinates = [
    ('Albany', 42.6526, -73.7562),
    ('Atlanta', 33.749, -84.388),
    ('Baltimore/Washington', 39.2904, -76.6122),
    ('Boise', 43.615, -116.2023),
    ('Boston', 42.3601, -71.0589),
    ('Buffalo/Rochester', 43.1566, -77.6088),
    ('California', 36.7783, -119.4179),
    ('Charlotte', 35.2271, -80.8431),
    ('Chicago', 41.8781, -87.6298),
    ('Cincinnati/Dayton', 39.1031, -84.512),
    ('Columbus', 39.9612, -82.9988),
    ('Dallas/Ft. Worth', 32.7767, -96.797),
    ('Denver', 39.7392, -104.9903),
    ('Detroit', 42.3314, -83.0458),
    ('Grand Rapids', 42.9634, -85.6681),
    ('Great Lakes', 45.0, -84.0),
    ('Harrisburg/Scranton', 40.2732, -76.8867),
    ('Hartford/Springfield', 41.7627, -72.6743),
    ('Houston', 29.7604, -95.3698),
    ('Indianapolis', 39.7684, -86.1581),
    ('Jacksonville', 30.3322, -81.6557),
    ('Las Vegas', 36.1699, -115.1398),
    ('Los Angeles', 34.0522, -118.2437),
    ('Louisville', 38.2527, -85.7585),
    ('Miami/Ft. Lauderdale', 25.7617, -80.1918),
    ('Midsouth', 35.1495, -90.0489),
    ('Nashville', 36.1627, -86.7816),
    ('New Orleans/Mobile', 29.9511, -90.0715),
    ('New York', 40.7128, -74.006),
    ('Northeast', 41.5, -75.0),
    ('Northern New England', 44.0, -71.5),
    ('Orlando', 28.5383, -81.3792),
    ('Philadelphia', 39.9526, -75.1652),
    ('Phoenix/Tucson', 33.4484, -112.074),
    ('Pittsburgh', 40.4406, -79.9959),
    ('Plains', 39.0119, -98.4842),
    ('Portland', 45.5152, -122.6784),
    ('Raleigh/Greensboro', 35.7796, -78.6382),
    ('Richmond/Norfolk', 37.5407, -77.436),
    ('Roanoke', 37.2709, -79.9414),
    ('Sacramento', 38.5816, -121.4944),
    ('San Diego', 32.7157, -117.1611),
    ('San Francisco', 37.7749, -122.4194),
    ('Seattle', 47.6062, -122.3321),
    ('South Carolina', 33.8361, -81.1637),
    ('South Central', 34.7465, -92.2896),
    ('Southeast', 32.3182, -86.9023),
    ('Spokane', 47.6588, -117.426),
    ('St. Louis', 38.627, -90.1994),
    ('Syracuse', 43.0481, -76.1474),
    ('Tampa', 27.9506, -82.4572),
    ('Total U.S.', 39.8283, -98.5795),
    ('West', 39.1178, -120.013),
    ('West Tex/New Mexico', 31.9686, -99.9018)
]


class GeoDistanceEncoder:
    def __init__(self, city_coordinates, reference_points):
        """
        Initialisiert den GeoDistanceEncoder mit den Städten und den Referenzpunkten.

        :param city_coordinates: Dictionary mit Städtenamen und deren (Latitude, Longitude)-Koordinaten
        :param reference_points: Dictionary mit Referenzpunkten und deren (Latitude, Longitude)-Koordinaten
        """
        self.city_coordinates = city_coordinates
        self.reference_points = reference_points

    def fit(self, X, y=None):
        # fit ist nicht notwendig, wird aber für Kompatibilität bereitgestellt.
        return self

    def transform(self, X):
        """
        Berechnet die geografischen Distanzen zu den Referenzpunkten. Falls die Spalten 'latitude' und 'longitude'
        nicht im Datensatz vorhanden sind, werden sie anhand der Stadtzuordnung berechnet.

        :param X: DataFrame mit der Spalte 'geography' (Städte) und optional 'latitude' und 'longitude'
        :return: DataFrame mit den berechneten Distanzen zu den Referenzpunkten
        """
        # Überprüfen, ob die Spalten 'latitude' und 'longitude' bereits vorhanden sind
        if 'latitude' not in X.columns or 'longitude' not in X.columns:
            # Falls die Spalten nicht existieren, die Koordinaten der Städte hinzufügen
            X['latitude'] = X['geography'].map(lambda x: self.city_coordinates.get(x, (None, None))[0])
            X['longitude'] = X['geography'].map(lambda x: self.city_coordinates.get(x, (None, None))[1])

        # Entfernen von Zeilen ohne gültige Koordinaten
        X = X.dropna(subset=['latitude', 'longitude'])

        # Berechnen der Entfernungen zu den Referenzpunkten
        for name, coords in self.reference_points.items():
            X[f'dist_to_{name}'] = X.apply(lambda row: geodesic((row['latitude'], row['longitude']), coords).kilometers, axis=1)

        return X

    def fit_transform(self, X, y=None):
        """
        Führt fit und transform direkt hintereinander aus.

        :param X: DataFrame mit der Spalte 'geography' (Städte) und optional 'latitude' und 'longitude'
        :return: DataFrame mit den berechneten Distanzen zu den Referenzpunkten
        """
        return self.fit(X).transform(X)

# Umwandeln der Liste in ein Dictionary für den Encoder
city_coordinates = {city: (lat, lon) for city, lat, lon in static_coordinates}

# Referenzpunkte (Kalifornien und Mexiko)
reference_points = {
    'california': (36.7783, -119.4179),
    'mexico': (23.6345, -102.5528)
}

# Anwendung des GeoDistanceEncoders
encoder = GeoDistanceEncoder(city_coordinates, reference_points)
df = encoder.fit_transform(df)

# Überprüfen der ersten Zeilen des Datensatzes
print(df[['geography', 'latitude', 'longitude', 'dist_to_california', 'dist_to_mexico']].head())
```

# Train-Test-Split

```python
# Funktion zur Aufteilung in Train und Test für einen gegebenen Datensatz
def split_train_test(df, target_column='average_price', train_year=2020):
    train_df = df[df['year'] < train_year]
    test_df = df[df['year'] >= train_year]
    X_train = train_df.drop(columns=[target_column])
    y_train = train_df[target_column]
    X_test = test_df.drop(columns=[target_column])
    y_test = test_df[target_column]
    return X_train, y_train, X_test, y_test

# Organic West
X_train_organic_west, y_train_organic_west, X_test_organic_west, y_test_organic_west = split_train_test(df_organic_west)
print("Organic West - Train-Daten:", X_train_organic_west.shape, y_train_organic_west.shape)
print("Organic West - Test-Daten:", X_test_organic_west.shape, y_test_organic_west.shape)

# Organic East
X_train_organic_east, y_train_organic_east, X_test_organic_east, y_test_organic_east = split_train_test(df_organic_east)
print("Organic East - Train-Daten:", X_train_organic_east.shape, y_train_organic_east.shape)
print("Organic East - Test-Daten:", X_test_organic_east.shape, y_test_organic_east.shape)

# Conventional West
X_train_conventional_west, y_train_conventional_west, X_test_conventional_west, y_test_conventional_west = split_train_test(df_conventional_west)
print("Conventional West - Train-Daten:", X_train_conventional_west.shape, y_train_conventional_west.shape)
print("Conventional West - Test-Daten:", X_test_conventional_west.shape, y_test_conventional_west.shape)

# Conventional East
X_train_conventional_east, y_train_conventional_east, X_test_conventional_east, y_test_conventional_east = split_train_test(df_conventional_east)
print("Conventional East - Train-Daten:", X_train_conventional_east.shape, y_train_conventional_east.shape)
print("Conventional East - Test-Daten:", X_test_conventional_east.shape, y_test_conventional_east.shape)
```

# Scaler

```python
# Import necessary libraries
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Liste der zu skalierenden Spalten
columns_to_scale = ['1/total_volume', '4046', '4225', '4770', 'small_bags', 'large_bags', 'xlarge_bags',
                    'dist_to_california', 'dist_to_mexico', 'lag_7', 'lag_30', 'moving_avg_30']

# Funktion zur Skalierung der Trainings- und Testdaten für einen Datensatz
def scale_data(X_train, X_test, columns):
    scaler = StandardScaler()
    X_train[columns] = scaler.fit_transform(X_train[columns])
    X_test[columns] = scaler.transform(X_test[columns])
    return X_train, X_test

# Skalierung für Organic West
X_train_organic_west, X_test_organic_west = scale_data(X_train_organic_west, X_test_organic_west, columns_to_scale)
print("Organic West - Skalierte Train-Daten:", X_train_organic_west.shape)
print("Organic West - Skalierte Test-Daten:", X_test_organic_west.shape)

# Skalierung für Organic East
X_train_organic_east, X_test_organic_east = scale_data(X_train_organic_east, X_test_organic_east, columns_to_scale)
print("Organic East - Skalierte Train-Daten:", X_train_organic_east.shape)
print("Organic East - Skalierte Test-Daten:", X_test_organic_east.shape)

# Skalierung für Conventional West
X_train_conventional_west, X_test_conventional_west = scale_data(X_train_conventional_west, X_test_conventional_west, columns_to_scale)
print("Conventional West - Skalierte Train-Daten:", X_train_conventional_west.shape)
print("Conventional West - Skalierte Test-Daten:", X_test_conventional_west.shape)

# Skalierung für Conventional East
X_train_conventional_east, X_test_conventional_east = scale_data(X_train_conventional_east, X_test_conventional_east, columns_to_scale)
print("Conventional East - Skalierte Train-Daten:", X_train_conventional_east.shape)
print("Conventional East - Skalierte Test-Daten:", X_test_conventional_east.shape)
```

```python
# Funktion zum Entfernen der angegebenen Spalte aus Trainings- und Testdaten
def drop_columns(X_train, X_test, columns):
    X_train = X_train.drop(columns=columns)
    X_test = X_test.drop(columns=columns)
    return X_train, X_test

# Entfernen der 'year' Spalte für Organic West
X_train_organic_west, X_test_organic_west = drop_columns(X_train_organic_west, X_test_organic_west, ['year'])

# Entfernen der 'year' Spalte für Organic East
X_train_organic_east, X_test_organic_east = drop_columns(X_train_organic_east, X_test_organic_east, ['year'])

# Entfernen der 'year' Spalte für Conventional West
X_train_conventional_west, X_test_conventional_west = drop_columns(X_train_conventional_west, X_test_conventional_west, ['year'])

# Entfernen der 'year' Spalte für Conventional East
X_train_conventional_east, X_test_conventional_east = drop_columns(X_train_conventional_east, X_test_conventional_east, ['year'])
```
