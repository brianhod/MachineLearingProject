```python
import matplotlib.pyplot as plt

# Data for MSE values
models = ['DecisionTree\nmax_depth = 3', 
          'with Validation\nmax_depth = 3', 
          'with GridSearch\nmax_depth = 15', 
          'RandomForest \nwith GridSearch\nmax_depth = 15',
          'RandomForest \nwith GridSearch \nmax_depth =5']

mse_train = [0.084746696, 0.097185548, 0.013580421, 0.007566076,0.056372584]
mse_test = [0.068799544, 0.039190584, 0.053592946, 0.041312375, 0.054931102]

# Create plot
plt.figure(figsize=(10, 6))

# Plotting the curves for training and test MSE
plt.plot(models, mse_train, label='MSE Train', marker='o', color='blue', linestyle='-', linewidth=2)
plt.plot(models, mse_test, label='MSE Test', marker='o', color='orange', linestyle='--', linewidth=2)

# Adding labels and title
plt.xlabel('DecisionTree')
plt.ylabel('MSE')
plt.title('MSE for Different Models on Train and Test Sets')

# Rotate x-axis labels to avoid overlapping
plt.xticks(rotation=45, ha='right')

# Adding a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
